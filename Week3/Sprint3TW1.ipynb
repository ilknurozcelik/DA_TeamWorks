### 7. Plot the avg tip by size (lineplot)

### 9. Make a scatter plot between tip and total_bill

```python
plt.scatter(tips["tip"], tips["total_bill"], linewidth = 0.001)
plt.xlabel("tip")
plt.ylabel("total_bill")
;
```

```python
sns.scatterplot(data=tips, x="tip", y="total_bill");
```

### 10. Set an additional dimension using size column

```python
plt.scatter(tips["tip"], tips["total_bill"],  c=tips["size"], cmap="coolwarm")
plt.colorbar()
plt.xlabel("tip")
plt.ylabel("total_bill")
;
```

```python
tips.plot.scatter(x = "tip", y = "total_bill", c = tips["size"], cmap = 'coolwarm', figsize=(10,6));
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.scatterplot(data=tips, x="tip", y="total_bill", hue="size", palette="coolwarm", legend="full");
```
